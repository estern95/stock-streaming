# ELT of TSLA Options

## Connect to kafka Topics

````
callsRaw = (spark.readStream
        .format("kafka")
        .option("kafka.bootstrap.servers", "34.136.146.75:29094")
        .option("subscribe", "TSLA_calls")
        .option("startingOffsets", "earliest") 
        .load())

putsRaw = (spark.readStream
        .format("kafka")
        .option("kafka.bootstrap.servers", "34.136.146.75:29094")
        .option("subscribe", "TSLA_puts")
        .option("startingOffsets", "earliest") 
        .load())

callsRaw.printSchema()
````

````
callsDF = callsRaw.selectExpr("CAST(value AS STRING)")
putsDF = putsRaw.selectExpr("CAST(value AS STRING)")
display(callsDF)
````

````
display(putsDF)
````

## Write to delta tables

````
(callsDF.writeStream
  .format("delta")
  .outputMode("append")
  .option("checkpointLocation", "/delta/events/_checkpoints/etl-from-json")
  .table("calls"))

````

````
(putsDF.writeStream
  .format("delta")
  .outputMode("append")
  .option("checkpointLocation", "/delta/events/_checkpoints/etl-from-json")
  .table("puts"))
````

## Perform type conversions

````
from pyspark.sql.functions import get_json_object, to_timestamp, from_unixtime
callsFormatted = callsDF.select(
  get_json_object(callsDF.value, "$.price").cast('double').alias('price'),
  get_json_object(callsDF.value, "$.contract").alias('contract'),
  to_timestamp(get_json_object(callsDF.value, "$.last_trade_dt"), 'yyyy-M-d h:ma z').alias('last_trade_dt'),
  get_json_object(callsDF.value, "$.strike").cast('double').alias('strike'),
  get_json_object(callsDF.value, "$.last_price").cast('double').alias('last_price'),
  get_json_object(callsDF.value, "$.bid").cast('double').alias('bid'),
  get_json_object(callsDF.value, "$.ask").cast('double').alias('ask'),
  get_json_object(callsDF.value, "$.change").alias('change'),
  get_json_object(callsDF.value, "$.volume").cast('int').alias('volume'),
  get_json_object(callsDF.value, "$.open_interest").cast('int').alias('open_interest'),
  get_json_object(callsDF.value, "$.implied_volatility").alias('implied_volatility'),
  to_timestamp(from_unixtime(get_json_object(callsDF.value, "$.option_expiration"))).alias('option_expiration'))
display(callsFormatted)         
````

````
putsFormatted = putsDF.select(
  get_json_object(putsDF.value, "$.price").cast('double').alias('price'),
  get_json_object(putsDF.value, "$.contract").alias('contract'),
  to_timestamp(get_json_object(putsDF.value, "$.last_trade_dt"), 'yyyy-M-d h:ma z').alias('last_trade_dt'),
  get_json_object(putsDF.value, "$.strike").cast('double').alias('strike'),
  get_json_object(putsDF.value, "$.last_price").cast('double').alias('last_price'),
  get_json_object(putsDF.value, "$.bid").cast('double').alias('bid'),
  get_json_object(putsDF.value, "$.ask").cast('double').alias('ask'),
  get_json_object(putsDF.value, "$.change").alias('change'),
  get_json_object(putsDF.value, "$.volume").cast('int').alias('volume'),
  get_json_object(putsDF.value, "$.open_interest").cast('int').alias('open_interest'),
  get_json_object(putsDF.value, "$.implied_volatility").alias('implied_volatility'),
  to_timestamp(from_unixtime(get_json_object(putsDF.value, "$.option_expiration"))).alias('option_expiration'))
display(putsFormatted) 
````
